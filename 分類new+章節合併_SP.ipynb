```python
import pandas as pd
import csv
```

```python
# readfile
name = 'SP1081_10_17_final'
data = pd.read_csv('./dataset/' + name + '.csv')
data.dropna(axis=0, how='any', inplace=True)
data.head(10)
```

```python
#增加章節資訊

data['章節']= ''
for i in range(len(data)):
    tmplist = data.iloc[i][4]
    if ('第一章' in tmplist or  '1-' in tmplist and '10-' not in tmplist and '11-' not in tmplist and '12-' not in tmplist):
        data.iat[i,5] = 'a'
    elif ('第二章' in tmplist or  '2-' in tmplist and '12-' not in tmplist ):
        data.iat[i,5] = 'b'
    elif ('第三章' in tmplist or  '3-' in tmplist ):
        data.iat[i,5] = 'c'
    elif ('第四章' in tmplist or  '4-' in tmplist ):
        data.iat[i,5] = 'd'
    elif ('第五章' in tmplist or  '5-' in tmplist ):
        data.iat[i,5] = 'e'
    elif ('第六章' in tmplist or  '6-' in tmplist ):
        data.iat[i,5] = 'f'
    elif ('第七章' in tmplist or  '7-' in tmplist ):
        data.iat[i,5] = 'g'
    elif ('第八章' in tmplist or  '8-' in tmplist ):
        data.iat[i,5] = 'h'
    elif ('第九章' in tmplist or  '9-' in tmplist ):
        data.iat[i,5] = 'i'
    elif ('第十章' in tmplist or  '10-' in tmplist):
        data.iat[i,5] = 'j'
    elif ('第十一章' in tmplist or  '11-' in tmplist):
        data.iat[i,5] = 'k'
    elif ('第十二章' in tmplist or  '12-' in tmplist):
        data.iat[i,5] = 'l'
data.head(10)
```

```python
#分類課程資訊

list_unknow = []
c = 0
for i in range(len(data)):
    tmplist = data.iloc[i][4]
    if ('教材大綱' in tmplist or '熱身區' in tmplist or '課程重要資訊說明' in tmplist or  '評分標準' in tmplist 
       or  '學前能力需求說明' in tmplist or  '學前能力需求說明投影片' == tmplist or  '章錄影檔' in tmplist
       or  '補充教材' == tmplist or  '組譯器(Assembler)' == tmplist or  '歷年考古題' == tmplist
       or  '參考資料' == tmplist or  '系統程式(遠距)' in tmplist or  '第三章 載入器及連結器(第6,7週)' in tmplist
       or  '第一章 系統程式簡介(第1,2週)' in tmplist or  '第二章 組譯器(Assembler)(第3,4,5週)' in tmplist):
        data.iat[i,4] = 'null'
        
        
    elif ('課前小叮嚀' in tmplist or  '課程內容說明' in tmplist or  '課程基本資料' in tmplist or  '學習活動及進度說明' in tmplist 
         or  '各單元內容及教學目標' in tmplist or  '教學評量及評鑑' in tmplist or  '評分標準投影片' in tmplist or  '評分標準錄影檔' in tmplist
         or  '學前能力需求說明' in tmplist ):
        data.iat[i,4] = '相關規定'
        
    elif ('投影片' in tmplist or ( '習題解答' in tmplist and '影片' not in tmplist )   
         or  '組譯器之實作Assembler' == tmplist or  '指令翻譯' == tmplist or  '課程結語' == tmplist ):
        data.iat[i,4] = '完整投影片'   

    elif ('影片' in tmplist or 'chap' in tmplist or '課程動畫' in tmplist ):
        data.iat[i,4] = '完整教材錄影'        
    
    elif ('學習回顧與整理' in tmplist):
        data.iat[i,4] = '課程相關疑難釋疑含相關規定'            
        
        
    #繳交作業 = null ， 因為作業使用實體課程資訊   
    elif ('作業' in tmplist ):
        data.iat[i,4] = 'null'  
        
    elif ('測驗' in tmplist or '小考' in tmplist or '牛刀小試' in tmplist or '期中考' in tmplist):
        data.iat[i,4] = '測驗'    

    elif ('課程重點整理' in tmplist ):
        data.iat[i,4] = '重點整理' 
        
    elif ('上課筆記' in tmplist or 'word檔' in tmplist or '如何撰寫編譯器' in tmplist or '參考資料' in tmplist
         or '書面資料' in tmplist or '如何撰寫組譯器' in tmplist or '解Token程式執行範例' in tmplist
         or '名詞解釋' in tmplist or 'Microsoft obj com exe格式' in tmplist or '解token_程式執行範例' in tmplist
         or 'sp_1' in tmplist or '分組討論紀錄表' in tmplist):
        data.iat[i,4] = '教材下載' 
        
    elif ('年' in tmplist ):
        data.iat[i,4] = '考古題'   
        
    elif ('s)' in tmplist ):
        data.iat[i,4] = '完整教材錄影'  
    
    elif ('?' in tmplist or '我學到哪些內容' in tmplist or '我的軟體開發經驗' in tmplist 
          or '我所知道的Parsing方法' in tmplist or 'complier' in tmplist or 'interpreter' in tmplist
         or 'SQL' in tmplist or '解釋' in tmplist or 'SQL' in tmplist or 'DBMS' in tmplist):
        data.iat[i,4] = '討論'
        

    else:
        print(data.iat[i,4])
        list_unknow.append((data.iat[i,4]))
        data.iat[i,4] = '未知'
        c = c+1

        
print('未知筆數 = ',c)
print(set(list_unknow))
```

```python
# 清除未知、null課程
data = data[data['title'] != '未知' ]
data = data[data['title'] != 'null' ]
data.to_csv(name + '_chap.csv',index = False)
print(name + '_chap')
```
